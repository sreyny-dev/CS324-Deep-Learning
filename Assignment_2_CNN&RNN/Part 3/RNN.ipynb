## Train Data with T=4

```python
import csv
import matplotlib.pyplot as plt

filename = 'results/T4.csv'
```

```python
    
def read_plot(filename, T):
    epochs = []
    train_acc = []
    val_acc = []
    train_loss = []
    val_loss = []
    
    with open(filename, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            epochs.append(int(row['Epoch']))
            train_acc.append(float(row['Train Acc']))
            val_acc.append(float(row['Val Acc']))
            train_loss.append(float(row['Train Loss']))
            val_loss.append(float(row['Val Loss']))
    
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_acc, label='Train Accuracy')
    plt.plot(epochs, val_acc, label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title(f'Training and Validation Accuracy T={T}')
    plt.legend()
    plt.grid()
    
    plt.subplot(1, 2, 2)
    plt.plot(epochs, train_loss, label='Train Loss')
    plt.plot(epochs, val_loss, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.grid()
    
    plt.tight_layout()
    plt.show()
```

```python
read_plot(filename, 4)
```

## T=5

```python
t5 = 'results/T5.csv'
read_plot(t5, 5)
```

## T=10

```python
t10 = 'results/T10.csv'
read_plot(t10, 10)
```
