# Task 1: Generate a dataset of points

```python
import perceptron 
mean_1, mean_2 = 2, 4
var_1, var_2 = 1, 1
x_train, y_train, x_test, y_test = perceptron.generate_data(mean_1, mean_2, var_1, var_2)
# Optional: Visualize the training and testing data
plt.title('Training Data')
plt.scatter(x_train[y_train == 1][:, 0], x_train[y_train == 1][:, 1], label='train_data1')
plt.scatter(x_train[y_train == -1][:, 0], x_train[y_train == -1][:, 1], label='train_data2')
plt.legend()
plt.show()

plt.title('Testing Data')
plt.scatter(x_test[y_test == 1][:, 0], x_test[y_test == 1][:, 1], label='test_data1')
plt.scatter(x_test[y_test == -1][:, 0], x_test[y_test == -1][:, 1], label='test_data2')
plt.legend()
plt.show()
```

### Test the accuracy of the testing data:

```python
from perceptron import Perceptron
perceptron = Perceptron(n_inputs=2)

training_accu, testing_accu, training_loss, testing_loss = perceptron.train(x_train, y_train, x_test, y_test)

training_accuracy = np.array(training_accu)
testing_accuracy  = np.array(testing_accu )
training_loss = np.array(training_loss)
testing_loss = np.array(testing_loss)


# Plot the accuracy vs. epoch
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot accuracy vs. epoch
axs[0].plot(training_accuracy, label='Training Accuracy')
axs[0].plot(testing_accuracy, label='Test Accuracy')
axs[0].set_title(f'Accuracy vs. Epoch (Mean=  {mean_1}, {mean_2}; Var = {var_1}, {var_2})')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Accuracy')
axs[0].legend()

# Plot loss vs. epoch
axs[1].plot(training_loss, label='Training Loss')
axs[1].plot(testing_loss, label='Test Loss')
axs[1].set_title(f'Loss vs. Epoch (Mean=  {mean_1}, {mean_2}; Var = {var_1}, {var_2})')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Loss')
axs[1].legend()

```

# Case 1. If the means of the two Gaussians are too close

```python
import perceptron
from perceptron import Perceptron
p = Perceptron(n_inputs=2)
#generate dataset
mean_1, mean_2 = 2, 2.1
var_1, var_2 = 1, 1
x_train, y_train, x_test, y_test = perceptron.generate_data(mean_1, mean_2, var_1, var_2)
training_accu, testing_accu, training_loss, testing_loss = p.train(x_train, y_train, x_test, y_test)

print(max(training_accu))
print(min(training_accu))
print(max(testing_accu))
print(min(testing_accu))

training_accuracy = np.array(training_accu)
testing_accuracy  = np.array(testing_accu )
training_loss = np.array(training_loss)
testing_loss = np.array(testing_loss)


# Plot the accuracy vs. epoch
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot accuracy vs. epoch
axs[0].plot(training_accuracy, label='Training Accuracy')
axs[0].plot(testing_accuracy, label='Test Accuracy')
axs[0].set_title(f'Accuracy vs. Epoch (Mean=  {mean_1}, {mean_2}; Var = {var_1}, {var_2})')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Accuracy')
axs[0].legend()

# Plot loss vs. epoch
axs[1].plot(training_loss, label='Training Loss')
axs[1].plot(testing_loss, label='Test Loss')
axs[1].set_title(f'Loss vs. Epoch (Mean=  {mean_1}, {mean_2}; Var = {var_1}, {var_2})')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Loss')
axs[1].legend()
    
```

# Analysis
When the means of X1 and X2 are too close, the training accuracy of the perceptron decreases because the data from both classes becomes harder to separate. The perceptron, being a linear classifier, struggles to find a clear decision boundary between overlapping points from the two classes. This overlap increases classification errors during training, resulting in lower accuracy.

In short, when the class distributions are too similar, the perceptron cannot effectively distinguish between them.

# Case 2: if their variance is too high


```python
import perceptron
from perceptron import Perceptron
p = Perceptron(n_inputs=2)
#generate dataset
mean_1, mean_2 = 2, 2.1
var_1, var_2 = 10, 10
x_train, y_train, x_test, y_test = perceptron.generate_data(mean_1, mean_2, var_1, var_2)
training_accu, testing_accu, training_loss, testing_loss = p.train(x_train, y_train, x_test, y_test)

print(max(training_accu))
print(min(training_accu))
print(max(testing_accu))
print(min(testing_accu))

training_accuracy = np.array(training_accu)
testing_accuracy  = np.array(testing_accu )
training_loss = np.array(training_loss)
testing_loss = np.array(testing_loss)


# Plot the accuracy vs. epoch
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot accuracy vs. epoch
axs[0].plot(training_accuracy, label='Training Accuracy')
axs[0].plot(testing_accuracy, label='Test Accuracy')
axs[0].set_title(f'Accuracy vs. Epoch (Mean=  {mean_1}, {mean_2}; Var = {var_1}, {var_2})')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('Accuracy')
axs[0].legend()

# Plot loss vs. epoch
axs[1].plot(training_loss, label='Training Loss')
axs[1].plot(testing_loss, label='Test Loss')
axs[1].set_title(f'Loss vs. Epoch (Mean=  {mean_1}, {mean_2}; Var = {var_1}, {var_2})')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Loss')
axs[1].legend()
    
```

# Anaylysis
With higher variance, the data points for X1 and X2 are more widely scattered, causing significant overlap between the two classes. As a result, the perceptron struggles to find a clean, distinct linear boundary between them. This increases the likelihood of misclassifying points that are near the overlapping regions, reducing both training and testing accuracy.
